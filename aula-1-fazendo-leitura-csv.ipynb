```python
dados_sus = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/Pandas/main/dados_sus.csv", sep=";", encoding="ISO-8859-1", skiprows=[0, 1, 2], skipfooter=9, engine="python")
dados_sus.head()
```

### Código acima explicado:
- `pd.read_csv`: Função do pandas para ler arquivos CSV.
- `"https://raw.githubusercontent.com/alura-cursos/Pandas/main/dados_sus.csv"`: URL do arquivo CSV a ser lido.
- `sep=";"`: Define o delimitador usado no arquivo CSV como ponto e vírgula, em vez do padrão vírgula.
- `encoding="ISO-8859-1"`: Define a codificação do arquivo como ISO-8859-1 (também conhecido como Latin-1), útil para caracteres especiais.
- `skiprows=[0, 1, 2]`: Pula as primeiras três linhas do arquivo CSV, que geralmente são cabeçalhos ou informações indesejadas.
- `skipfooter=9`: Pula as últimas nove linhas do arquivo CSV, excluindo informações indesejadas no final.
- `engine="python"`: Especifica o uso do motor Python para a leitura do arquivo, necessário ao usar o parâmetro `skipfooter`.
