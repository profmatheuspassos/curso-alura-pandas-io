##### Usando Google Planilhas

Endereço da planilha: https://docs.google.com/spreadsheets/d/1q3DaGrBPxhEBGx5ECDd2TyvkgW8A88ZVIIUvkMkh7RE/edit?usp=sharing. O ID da planilha está entre o "d/" e o "/edit"

```python
sheet_id = "1q3DaGrBPxhEBGx5ECDd2TyvkgW8A88ZVIIUvkMkh7RE"
```

```python
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet"
# Parâmetros gviz/tq: permitem o acesso à API de visualização da Google
# Parâmetros ?tqx=out:csv: traz o arquivo como CSV
# Parâmetros &sheet: traz todos os dados da planilha
```

```python
dados_co2_sheets = pd.read_csv(url)
```

```python
dados_co2_sheets.head()
```
